```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("FinalOutput_orden.csv")
```

```python
dataset
```

```python
dataset=dataset.drop("S.No",axis=1)
```

```python
dataset.sort_values(["BP", "AT", "PM10", "RH", "PM2.5", "WD", "WS", "NH3", "CO", "NO", "NO2", "NOx", "SO2", "Ozone"], axis=0, inplace=True)
```

```python
dataset
```

```python
dataset.groupby(["Label"]).size()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset["Label"]=label_encoder.fit_transform(dataset["Label"])
dataset["Label"].unique()
```

```python
indep = dataset[["BP", "AT","PM10","AT", "RH", "PM2.5", "WD", "WS", "NH3", "CO", "NO", "NO2", "NOx", "SO2", "Ozone",'AQI Simulink ']];
dep = dataset[["Label"]];
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]} 
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1') 
grid.fit(X_train, y_train) 
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(X_test) 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions) 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

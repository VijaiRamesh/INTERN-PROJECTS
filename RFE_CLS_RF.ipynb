```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("resumes_development.csv")
```

```python
data
```

```python
data.info()
```

```python
data=data.drop("Unnamed: 0", axis=1)
```

```python
data.groupby(["Interview"]).size()
```

```python
df_grouped_by = data.groupby(['Interview'])
df_balanced = df_grouped_by.apply(lambda x: x.sample(df_grouped_by.size().min()).reset_index(drop=True))
data = df_balanced.droplevel(['Interview'])
```

```python
data.groupby(["Interview"]).size()
```

```python
df = data.copy()
```

```python
X=data.drop("Interview",axis=1)
y=df["Interview"]
```

```python
X
```

```python
y
```

```python
#X = pd.DataFrame(data = data.data, columns=data.feature_names)
```

```python
#y=data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
X_train.shape, X_test.shape
```

```python
sel = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1))
sel.fit(X_train, y_train)
sel.get_support()
```

```python
X_train.columns
```

```python
features = X_train.columns[sel.get_support()]
```

```python
features
```

```python
len(features)
```

```python
X_train_rfc = sel.transform(X_train)
X_test_rfc = sel.transform(X_test)
```

```python
np.mean(sel.estimator_.feature_importances_)
```

```python
sel.estimator_.feature_importances_
```

```python
def run_randomForest(X_train, X_test, y_train, y_test):
    clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print('Accuracy: ', accuracy_score(y_test,y_pred))
```

```python
%%time
run_randomForest(X_train_rfc, X_test_rfc, y_train, y_test)
```

```python
%%time
run_randomForest(X_train, X_test, y_train, y_test)
```

# Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE
sel = RFE(RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1), n_features_to_select = 15)
sel.fit(X_train, y_train)
```

```python
sel.get_support()
```

```python
features = X_train.columns[sel.get_support()]
```

```python
features
```

```python
len(features)
```

```python
X_train_rfe = sel.transform(X_train)
X_test_rfe = sel.transform(X_test)
```

```python
%%time
run_randomForest(X_train_rfe, X_test_rfe, y_train, y_test)
```

```python
%%time
run_randomForest(X_train, X_test, y_train, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
sel = RFE(AdaBoostClassifier(n_estimators=100, random_state=0), n_features_to_select = 15)
sel.fit(X_train, y_train)
```

```python
for index in range(1,11):
    from sklearn.ensemble import RandomForestClassifier
    sel = RFE(RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0), n_features_to_select = index)
    sel.fit(X_train, y_train)
    X_train_rfe = sel.transform(X_train)
    X_test_rfe = sel.transform(X_test)
    features = X_train.columns[sel.get_support()]
    print('Selected Feature: ',index,features)
    run_randomForest(X_train_rfe, X_test_rfe, y_train, y_test)
    print()
```
